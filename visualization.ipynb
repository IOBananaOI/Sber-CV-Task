```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "results/"
```

```python
names = ["matcha-base", "matcha-chartqa", "matcha-plotqa"]
```

```python
plot_settings = {
    "matcha-base" : {
        "marker" : "o",
        "color" : "purple"
    },
    "matcha-chartqa" : {
        "marker" : "^",
        "color" : "green"
    },
    "matcha-plotqa" : {
        "marker" : "*",
        "color" : "red"
    }
}
```

Loss (Train)

```python
loss_train = {}
for name in names:
    loss_train[name] = pd.read_csv(f"{path}{name}/Loss (Train).csv", header=None)[2].values[:600]
```

```python
plt.figure(figsize=(16, 7))
plt.title("Loss (Train) comparison")
for k, v in loss_train.items():
    plt.plot(v[::15], color=plot_settings[k]['color'], marker=plot_settings[k]['marker'], label=k)

plt.xlabel("Step")
plt.ylabel("Loss (Train)")
plt.xticks(np.arange(0, 40), np.arange(0, 600, 15))
plt.legend()
plt.show()
```

Loss (Test)

```python
for metric in ["Loss (Test)", 'ANLS', "EM", "Relaxed correctness"]:
    data = {}
    for name in names:
        data[name] = pd.read_csv(f"{path}{name}/{metric}.csv", header=None)[2].values
    
    plt.figure(figsize=(10, 6))
    plt.title(f"{metric} comparison")
    for k, v in data.items():
        plt.plot(v, color=plot_settings[k]['color'], marker=plot_settings[k]['marker'], label=k)

    plt.xlabel("Step")
    plt.ylabel(f"{metric}")
    plt.xticks(np.arange(0, 6), np.arange(100, 700, 100))
    plt.legend()
    plt.show()
```
